```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
train = pd.read_json('/content/drive/MyDrive/ML/ML1/data/train.json')[['bathrooms', 'bedrooms', 'price']]
test = pd.read_json('/content/drive/MyDrive/ML/ML1/data/test.json')[['bathrooms', 'bedrooms', 'price']]
```

```python
poly = PolynomialFeatures(10)

pd.DataFrame(poly.fit_transform(train))
```

```python
pd.DataFrame(poly.fit_transform(test))
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import (
    mean_absolute_error,
    root_mean_squared_error
)
```

```python
result_MAE = pd.DataFrame(columns=['model', 'train', 'test'])
result_RMSE = pd.DataFrame(columns=['model', 'train', 'test'])
```

```python
X_train = train[['bathrooms', 'bedrooms']]
X_test = test[['bathrooms', 'bedrooms']]
y_train = train['price']
y_test = test['price']
```

```python
lgs = LinearRegression()
lgs.fit(X_train, y_train)
```

```python
mae_test = mean_absolute_error(y_test, lgs.predict(X_test))
mae_train = mean_absolute_error(y_train, lgs.predict(X_train))
rmse_test = root_mean_squared_error(y_test, lgs.predict(X_test))
rmse_train = root_mean_squared_error(y_train, lgs.predict(X_train))
```

```python
result_MAE.loc[len(result_MAE)] = {'model':'linear_regression', 'train':mae_train, 'test':mae_test}
result_RMSE.loc[len(result_RMSE)] = {'model':'linear_regression', 'train':rmse_train, 'test':rmse_test}
```

```python
dt = DecisionTreeRegressor(random_state=21)

dt.fit(X_train, y_train)
```

```python
mae_test = mean_absolute_error(y_test, dt.predict(X_test))
mae_train = mean_absolute_error(y_train, dt.predict(X_train))
rmse_test = root_mean_squared_error(y_test, dt.predict(X_test))
rmse_train = root_mean_squared_error(y_train, dt.predict(X_train))
```

```python
result_MAE.loc[len(result_MAE)] = {'model':'decision_tree', 'train':mae_train, 'test':mae_test}
result_RMSE.loc[len(result_RMSE)] = {'model':'decision_tree', 'train':rmse_train, 'test':rmse_test}
```

```python
train['mean'] = train['price'].mean()
train['median'] = train['price'].median()
test['mean'] = test['price'].mean()
test['median'] = test['price'].median()
```

```python
mae_test = mean_absolute_error(y_test, test['mean'])
mae_train = mean_absolute_error(y_train, train['mean'])
rmse_test = root_mean_squared_error(y_test, test['mean'])
rmse_train = root_mean_squared_error(y_train, train['mean'])
```

```python
result_MAE.loc[len(result_MAE)] = {'model':'naive_mean', 'train':mae_train, 'test':mae_test}
result_RMSE.loc[len(result_RMSE)] = {'model':'naive_mean', 'train':rmse_train, 'test':rmse_test}
```

```python
mae_test = mean_absolute_error(y_test, test['median'])
mae_train = mean_absolute_error(y_train, train['median'])
rmse_test = root_mean_squared_error(y_test, test['median'])
rmse_train = root_mean_squared_error(y_train, train['median'])
```

```python
result_MAE.loc[len(result_MAE)] = {'model':'naive_median', 'train':mae_train, 'test':mae_test}
result_RMSE.loc[len(result_RMSE)] = {'model':'naive_median', 'train':rmse_train, 'test':rmse_test}
```
